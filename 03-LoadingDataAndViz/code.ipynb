```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
import numpy as np
import plotly.express as px

iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
pen = sns.load_dataset('penguins')

x = np.linspace(-np.pi, np.pi, 50)
y1 = np.cos(x)
y2 = np.sin(x)
```

```python
## functional approach
fig = plt.figure(figsize=(2,2))
plt.plot(x, y1, label='cosine')
plt.plot(x, y2, label='sine')
plt.xlabel('values')
plt.ylabel('function')
plt.title('My Plot')
plt.legend(loc=(1.1,0))
plt.show()
```

```python
fig = plt.figure(figsize=(3,3))
plt.scatter(iris['sepal_length'], iris['sepal_width']);
```

```python
# OOP approach
fig, ax = plt.subplots(figsize=(3,3))
ax.plot(x, y1, label='cosine')
ax.plot(x, y2, label='sine')
ax.set_xlabel('value')
ax.set_ylabel('function')
ax.set_title('My Plot');
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,3))
ax[0].plot(x, y1, label='cosine')
ax[1].plot(x, y2, label='sine')
```

```python
fig = plt.figure(figsize=(3,3))
plt.scatter(data=iris, x='sepal_length', y='sepal_width', c='orange')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
fig.savefig('scatter.png')
```

```python
fig = plt.figure(figsize=(3,3))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.xlabel('Sepal Length')
plt.legend(loc=(1.1, 0));
```

```python
fig = px.scatter(data_frame=iris, x='sepal_length', y='sepal_width', color='species', 
                 hover_data=['petal_length', 'petal_width'])
fig.update_layout(width=500, height=500)
fig.write_html('scatter.html')
```
